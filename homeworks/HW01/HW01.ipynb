## **Примеры использования NumPy**

### Импорт NumPy в общепринятом виде

```python
import numpy as np
```

### Создание массивов в NumPy

Простое создание массива из классического списка Python

```python
marks = np.array([4,5,5,3,2,5,4,4,3,5])
marks
```

Создание многомерного массива **n** x **m**, заполненного _нулями_ 

```python
n,m = 3,5
weights = np.zeros((n,m))
weights
```

Такое же создание многомерного массива **n**x**m**, заполненного _единицами_

```python
n,m = 2,3
weights = np.ones((n,m))
weights
```

Создание одномерного массива с указанием последовательности чисел внутри (диапазона)

```python
pages_for_load = np.arange(3,11)
pages_for_load
```

### Демонстрация примеров атрибутов для массива

Создаем для примера массив нулевых чисел 4 на 5

```python
weights = np.zeros((4,5))
```

С помощью атрибута **shape** можно вывести кортеж размеров массива по каждой оси

```python
weights.shape
```

С помощью атрибута **ndim** можно вывести количество осей в массиве (одномерный - 1, двумерный - 2, ...)

Атрибут **dtype** поможет узнать какой тип данных хранит массив

```python
weights.dtype
```

### Индексирование и срезы

Создаем двумерный массив чисел для примеров

```python
nums = np.arange(1,13).reshape(4,3) # используем функцию reshape для преобразования массива из одномерного в многомерный
nums
```

С помощью индексирования мы можем получить отдельные элементы массива

```python
nums[1,2]
```

Также с помощью индексов можно делать срезы конкретных нужных частей массива

```python
print(nums[1:3]) # 2 и 3 строки массива
print("-"*10)
print(nums[:,1]) # 2 столбец массива
print("-"*10)
print(nums[1:3,1:3]) # выбор среза 2х2
```

### Примеры векторных операций и базовых статистик

Массивы NumPy поддерживают очень удобные поэлементные операции, ниже приведу несколько примеров

Создание массива со статистикой роста людей, и прибавление к каждому значению 5

```python
people_heights = np.array([168,192,185,157,178,169,173])

people_heights+5
```

Можно взять корень из каждого значения

```python
people_heights**(1/2)
```

Также перевод каждого значения из сантиметров в метры

```python
people_heights/100
```

Создание статистики количества людей с ростом больше 175

```python
people_heights>175
```

Для дальнейшей работы нам часто приходится использовать статистику по используемым массивам

Для нахождения среднего значения будем использовать функцию mean

```python
people_heights.mean()
```

Классический поиск максимального max() и минимального min()

```python
print(people_heights.max())
print(people_heights.min())
```

Для примера хотелось бы еще привести использование функции стандартного отклонения std() (можно использовать как для всего массива так и по осям)

```python
people_heights.std()
```
